### Streaming and Batching

## Streaming
Most models can stream their output content while it is being generated. By displaying output progressively, streaming significatly improves user experience, particularly for longer responses. Calling stream() returns an iterator that yields the output chunks as they are produced. You can use a loop to process each chunk in real-time.

```python
import os
from dotenv import load_dotenv
load_dotenv()
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
for chunk in model.stream("What is artification intelligence?"):
    print(chunk.text)
```

### Batching
Batching is a collection of independent requests to a model can significantly can improve performance and reduce costs, as the processing can be done in parallel.

```python
responses = model.batch([
    "what is color of sky",
    "how to build llm",
    "what project should I develop that should be meaningful using generative ai and fullstack development using react/react native at frontend and python fastapi as backend"
    ],
    config ={
        'max_concurrency':5, # limit to 5 parallel calls.
    }
)

for response in responses:
    print(response.content)
```
