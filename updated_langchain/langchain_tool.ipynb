### Tool
Model can request to call tools that can perform task such as fetching data from the database, searching the web or running a code. Tools are pairings of:

1. A schema, naming of the tool, a description, and/or argument definitions(often a json schema).
2. A function or coroutine to execute.

```python
import os
from dotenv import load_dotenv
load_dotenv()
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.tools import tool

@tool
def get_weather(location:str) -> str:
    """Get weather for the location"""
    return f"The weather of {location} is cloudy and misty."


model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

model_with_tools = model.bind_tools([get_weather])
response = model_with_tools.invoke("What is the weather of boston?")
print(response)
for tool_call in response.tool_calls:
    print("Tool Name: ",tool_call["name"])
    print(f"Tool Content: {tool_call["args"]}")
```
